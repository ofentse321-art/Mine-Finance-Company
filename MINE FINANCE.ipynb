```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv(r"C:/Users/ofentse/Documents/mining_company_finance_dataset.csv")
df

#Filterig 

#Cash_Flow_Operations_USD > 8000000,500000

First_Task = df["Cash_Flow_Operations_USD"] > 80000000,500000
print("Cash flow operations greater than 80000000,500000 = ", First_Task)

#Royalties_Paid_USD < 150000000,2500000

Second_Task = df["Royalties_Paid_USD"] < 150000000,2500000
print(Second_Task)

#TotaL_Employees > 1500 & Budgeted-Revenue_USD > 51000000,800000

Third_Task = (df["Total_Employees"] > 1500) & (df["Budgeted_Revenue_USD"] > 51000000.800000)
print(Third_Task)

#Debt_to_Equity_Ratio

Fourth_Task = df["Debt_to_Equity_Ratio"]
print(Fourth_Task)

#Current_Ratio > 1,9000000000 | Debt_to_Equity_Ratio > 1,800000000

Fifth_Task = (df["Current_Ratio"] > 1.900000000) | (df["Debt_to_Equity_Ratio"] > 1.800000000)
print(Fifth_Task)

#Statistics

#Market_Share_Percet Average

ave = df["Market_Share_Percent"].mean()
print("Average of the Market_Share_Percent = ", ave)

#Insurance_Premiums_USD range

rang = df["Insurance_Premiums_USD"].max() - df["Insurance_Premiums_USD"].min()
print("The range of Insurance Premiums = ", rang)

#Return_on_Assets_Percent Minimum

least = df["Return_on_Assets_Percent"].min()
print("Minimum amount of return assets percentage = ", least)

#Revenue_USD Sum

total = df["Revenue_USD"].sum()
print("Sum of reenue = ", total)

#Predictios 

P = df[["Revenue_USD"]]
p = df["Operating_Expenses_USD"]

P_train, P_test, y_train, y_test = train_test_split(P, p, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(P, p)

prediction_1 = model.predict([[50000]])
prediction_2 = model.predict([[27000]])

print("The difference of the expenses in the company 1 : ", prediction_1)
print("The differece of the expense in the company 2 : ", prediction_2)

plt.scatter(P, p, color="red")
plt.title("Difference of the expenses in the company")
plt.xlabel("Total Money")
plt.ylabel("Totsl money used")
plt.show()

```
